```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

incols=[]
for i in range(784):
    incols.append("pixel"+str(i))
outcols=["label"]

traindf = pd.read_csv('../data/train.csv')
trainX = traindf.as_matrix(columns=incols)
trainy = traindf.as_matrix(columns=outcols)
```

```python
# Draw some of the images

num_rows = 3
num_cols = 6
N = num_rows*num_cols

fig = plt.figure()
gs = matplotlib.gridspec.GridSpec(num_rows, num_cols, wspace=0.0)
ax = [plt.subplot(gs[i]) for i in range(N)]

print trainy.ravel()[0:N]

for i in range(N):
    img = np.array(trainX[i]).reshape(28,28)
    ax[i].imshow(img)
    ax[i].axis('off')
```

```python
import sklearn.ensemble as ensemble

rfc = ensemble.RandomForestClassifier()
rfc.fit(trainX,trainy.ravel())
```

```python
testdf = pd.read_csv('../data/test.csv')
testX = testdf.as_matrix(columns=incols)

predictions = pd.DataFrame(columns=["ImageId","Label"])
predictions["ImageId"] = range(1,28001)
predictions["Label"] = rfc.predict(testX)
predictions.to_csv('../data/rf_predictions.csv', index=False)
```

```python

num_rows = 3
num_cols = 6
N = num_rows*num_cols

# Let's see some of the predictions

print predictions[0:N][["Label"]].as_matrix().ravel()

# Draw the corresponding images to compare

fig2 = plt.figure()
gs2 = matplotlib.gridspec.GridSpec(num_rows, num_cols, wspace=0.0)
ax2 = [plt.subplot(gs2[i]) for i in range(N)]

for i in range(N):
    img = np.array(testX[i]).reshape(28,28)
    ax2[i].imshow(img)
    ax2[i].axis('off')
```

```python
from sklearn.model_selection import cross_val_score

#print len(trainX[0:1000]), len(trainy[0:1000])
cross_val_score(rfc, trainX, trainy.ravel(), scoring='accuracy')
```
